```python
from supabase import Client, create_client
from tqdm import tqdm
from dotenv import load_dotenv
import os
load_dotenv()

# --- Configuración de Supabase ---
SUPABASE_URL = os.environ.get("SUPABASE_URL")
SUPABASE_SERVICE_KEY = os.environ.get("SUPABASE_SERVICE_KEY")

if not SUPABASE_URL or not SUPABASE_SERVICE_KEY:
    print("Error: Faltan las variables de entorno SUPABASE_URL o SUPABASE_SERVICE_KEY.")
    exit()

supabase: Client = create_client(SUPABASE_URL, SUPABASE_SERVICE_KEY)
# ---------------------------------
```

```python
supabase.rpc("select * from homilias").execute()
```

```python
from sqlalchemy import create_engine, text
import os
from dotenv import load_dotenv

# Cargar variables de entorno
load_dotenv()

# URL de conexión, del panel de Supabase:
# Ejemplo: postgresql://postgres:<PASSWORD>@db.<project>.supabase.co:5432/postgres
SUPABASE_URL = os.getenv("SUPABASE_URL")

# Crear el engine de SQLAlchemy
engine = create_engine(SUPABASE_URL)

# Tu consulta SQL
query = """
SELECT *
FROM homilias_traducciones
INNER JOIN homilias
    ON homilias_traducciones.homilia_id = homilias.id
WHERE homilias_traducciones.id = '4ee3030f-7323-4247-817f-e3efca3574b4'
LIMIT 10;
"""

# Ejecutar y mostrar resultados
with engine.connect() as conn:
    result = conn.execute(text(query))
    for row in result.mappings():  # mappings() → devuelve dicts con nombres de columnas
        print(dict(row))
```

```python
import psycopg2
from dotenv import load_dotenv
import os

# Load environment variables from .env
load_dotenv()

# Fetch variables
USER = os.getenv("user")
PASSWORD = os.getenv("password")
HOST = os.getenv("host")
PORT = os.getenv("port")
DATABASE_URL = os.getenv("DATABASE_URL")

# Connect to the database
try:
    connection = psycopg2.connect(
        database=DATABASE_URL
    )
    print("Connection successful!")
    
    # Create a cursor to execute SQL queries
    cursor = connection.cursor()
    
    # Example query
    cursor.execute("SELECT NOW();")
    result = cursor.fetchone()
    print("Current Time:", result)

    # Close the cursor and connection
    cursor.close()
    connection.close()
    print("Connection closed.")

except Exception as e:
    print(f"Failed to connect: {e}")
```

```python
from sqlalchemy import create_engine
# from sqlalchemy.pool import NullPool
from dotenv import load_dotenv
import os

# Load environment variables from .env
load_dotenv()

# Fetch variables
USER = os.getenv("user")
PASSWORD = os.getenv("password")
HOST = os.getenv("host")
PORT = os.getenv("port")
DBNAME = os.getenv("dbname")

# Construct the SQLAlchemy connection string
DATABASE_URL = f"postgresql+psycopg2://{USER}:{PASSWORD}@{HOST}:{PORT}/{DBNAME}?sslmode=require"

# Create the SQLAlchemy engine
engine = create_engine(DATABASE_URL)


# Test the connection
try:
    with engine.connect() as connection:
        print("Connection successful!")
except Exception as e:
    print(f"Failed to connect: {e}")
```
